```python
import json
import random

# My API Key
api_key = "0AvI8UdqhTQELbvWCOTmagtt"

# Companies for dataset
companies = ["Capri Holdings Ltd", "Nordstrom", "Macys", "Carrefour SA", "Foot Locker Inc", "Target", "American Eagle Outfitters","PepsiCo Inc",
    "Hermes International", "Urban Outfitters"]

# 3 ESG metrics
metrics = [
    ("cdp_scope_1_emissions", "TCo2e"),       # Greenhouse gas emissions
    ("renewable_energy_percentage", "%"),     # Renewable energy %
    ("health_safety_incidents", "incidents")  # Workplace incidents
]

records = []
# Loop through companies and metrics
for comp in companies:
    for metric, unit in metrics:

        # Generate realistic values
        if unit == "TCo2e":
            val = random.randint(1000, 5000000)
        elif unit == "%":
            val = round(random.uniform(10, 100), 1)
        else:
            val = random.randint(0, 500)
            
        # Build record in Briink format
        rec = {
            "input": {
                "question": f"{metric.replace('_', ' ')}?",
                "custom_id": metric,
                "data_type": "Metric",
                "company": comp,
                "file_metas": [
                    {
                        "api_key_used": api_key
                    }
                ]
            },
            "reference_output": {
                "answer": f"{val} {unit}",
                "source_documents": [
                    {
                        "api_key_used": api_key
                    }
                ]
            },
            "structured_data": [
                {
                    "unit": unit,
                    "value": val,
                    "time_period": "2022"
                }
            ]
        }

        records.append(rec)

# Save to JSON file
with open("briink_wikirate_dataset.json", "w") as f:
    json.dump(records, f, indent=4)

print("Total records:", len(records))
```

```python
import json
import pandas as pd

# Load the Briink dataset
with open("briink_wikirate_dataset.json", "r") as f:
    data = json.load(f)

rows = []

# Extract key information in a human-friendly format
for record in data:
    company = record["input"]["company"]
    metric = record["input"]["custom_id"]
    question = record["input"]["question"]
    answer = record["reference_output"]["answer"]

    # Structured data (numeric, unit, year)
    structured = record.get("structured_data", [{}])[0]
    value = structured.get("value", "")
    unit = structured.get("unit", "")
    time_period = structured.get("time_period", "")

    # Build a clear row for humans and ML clarity
    rows.append({
        "Company": company,
        "Metric": metric,
        "Question": question,
        "Answer": answer,
        "Value": value,
        "Unit": unit,
        "Year": time_period
    })

# Convert to DataFrame
df = pd.DataFrame(rows)

# Display nicely
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
print(df)

# Optionally, save to CSV for easy review
df.to_csv("briink_wikirate_dataset_humanized.csv", index=False)
```
